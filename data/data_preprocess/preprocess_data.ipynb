# newshop

```python
df_1 = pd.read_csv('newshop.csv')
df_1.head(10)
```

```python
df_1.isna().sum()
```

```python
df_1 = df_1.dropna(thresh=2)
df_1 = df_1.dropna(subset=['name'])
```

```python
df_1["author_name"].fillna( method ='ffill', inplace = True)
df_1["cover_type"].fillna( method ='ffill', inplace = True)
df_1["num_page"].fillna( method ='ffill', inplace = True)
df_1["size"].fillna( method ='ffill', inplace = True)
df_1["publisher"].fillna( method ='ffill', inplace = True)
df_1["release_year"].fillna( method ='ffill', inplace = True)
```

```python
df_1['image_url'] = df_1['image_url'].fillna('No image_url')
```

```python
df_1.info()
```

```python
df_1 = df_1.reset_index()
```

```python
df_1 = df_1.drop(columns=['index'])
df_1
```

```python
df_1.to_csv('newshop_.csv')
```

```python
df_1 = df_1.drop(columns = ['_id'])
```

# tiki

```python
df_3 = pd.read_csv('tiki.csv')
df_3.head(10)
```

```python
df_3.isna().sum()
```

```python
df_3 = df_3.drop(columns=['translator'])
```

```python
df_3["author"].fillna( method ='ffill', inplace = True)
df_3["cover_type"].fillna( method ='ffill', inplace = True)
df_3["num_page"].fillna( method ='ffill', inplace = True)
df_3["size"].fillna( method ='ffill', inplace = True)
df_3["publisher"].fillna( method ='ffill', inplace = True)
df_3["year"].fillna( method ='ffill', inplace = True)
df_3["provider"].fillna( method ='ffill', inplace = True)
```

```python
df_3 = df_3.reset_index()
df_3 = df_3.drop(columns=['index'])
df_3
```

```python
df_3 = df_3.drop(columns = ['_id'])
```
